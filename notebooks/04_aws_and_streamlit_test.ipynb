## 📘 Streamlit Reference Guide

---

### What is Streamlit?
Streamlit is an open-source Python library that turns Python scripts into **interactive web apps** — no HTML, CSS, or JavaScript needed. Built specifically for data scientists and ML engineers.

---

### Why Use Streamlit?
| Benefit | Details |
|---|---|
| **Fast Development** | Write apps in pure Python |
| **Rapid Prototyping** | Share dashboards with stakeholders quickly |
| **Built for Data** | Native support for DataFrames, charts, metrics |
| **Hot Reloading** | Auto-refreshes on file save during development |
| **Easy Deployment** | Streamlit Cloud (free), or any server |

---

### How to Use Streamlit

**Run the app:**
```bash
streamlit run app/main.py
```

**Display elements:**
```python
st.title("My App")          # Large title
st.header("Section")        # Section header
st.write("Hello!")          # Text or data
st.dataframe(df)            # Interactive table
st.metric("Revenue", "$92", "+7%")  # KPI card
```

**Input widgets:**
```python
name = st.text_input("Your name")
choice = st.selectbox("Pick one", ["A", "B", "C"])
value = st.slider("Select range", 0, 100)
clicked = st.button("Run Analysis")
```

**Layout:**
```python
col1, col2 = st.columns(2)       # Side-by-side columns
with col1:
    st.write("Left side")
st.sidebar.header("Navigation")  # Sidebar panel
```

**Caching (important for performance):**
```python
@st.cache_data
def load_data():
    return pd.read_csv("data.csv")   # Only runs once, cached after
```

**Session State (persist values across reruns):**
```python
if "count" not in st.session_state:
    st.session_state["count"] = 0
st.session_state["count"] += 1
```

## 📘 FastAPI Reference Guide

The project outline offers **FastAPI** as an alternative to Streamlit for exposing model results as a callable API service.

---

### What is FastAPI?
FastAPI is a modern Python web framework for building **REST APIs** — it exposes your analysis results as HTTP endpoints that other systems (front-ends, mobile apps, dashboards) can call programmatically.

---

### Streamlit vs FastAPI — When to Use Which?

| | Streamlit | FastAPI |
|---|---|---|
| **Best for** | Interactive dashboards, demos | Backend APIs, automated pipelines |
| **Output** | Visual web page | JSON responses |
| **Users** | Data analysts, non-technical stakeholders | Developers, other systems |
| **This project** | Dashboard demo (`app/main.py`) | API to serve model outputs |
| **Example** | "Show me the RFM chart" | `GET /rfm/segments` → returns JSON |

---

### How to Use FastAPI

**Install:**
```bash
pip install fastapi uvicorn
```

**Run the API server:**
```bash
uvicorn app.api:app --reload --port 8000
```
